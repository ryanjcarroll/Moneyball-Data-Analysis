# Introduction

This analysis uses the 'Moneyball' dataset, which was famously used by Oakland Athletics GM Billy Beane and analyst Paul DePodesta in the early 2000s to revolutionize MLB player evaluation.  In this project, we will attempt to reproduce some of the conclusions by the Athletics club about what drives a team to reach the playoffs.

```python
data = pd.read_csv('baseball.csv')
```

This dataset contains data from 1962-2012. To accurately get a picture of what the 2001-2002 Athletics would have used, we will throw out all the data which is dated past the 2001 season.

```python
teams = data[data['Year']<2002]

teams.head()
```

## Making the Playoffs

Beane and Podesta were interested in dissecting what drives a team to success.  In the MLB, the top teams are placed into a playoff bracket after the conclusion of the regular season, with the winner emerging as the World Series champion.  During this postseason, teams compete in 5-game and 7-game series against one another to determine who advances.  These short series, compared to the 162-game regular season, introduce a high amount of variability and can be quite hard to predict or build around.

With this in mind, Beane and Podesta decided to focus their efforts on building a team which could make the playoffs as their primary goal.  They would then hope for the die to roll in their favor in the postseason games.  

One important piece of this analysis is determining the number of games a team needs to win to give themselves a good chance to make the playoffs.

```python
teams[teams['Playoffs']==1]['W'].mean()
```

We can see that the average playoff team in the dataset won between 95 and 96 games.

To further visualize the variation in playoff team wins, we create the categorical plot below, where the orange dots represent playoff teams and the blue dots represent non-playoff teams.  The vertical black line represents the average number of wins by playoff teams.

```python
sns.catplot(y='Team', x='W', data=teams, jitter=False, hue='Playoffs', height=8)

plt.axvline(teams[teams['Playoffs']==1]['W'].mean(), color='black')
plt.title('Games Won by Playoff and Non-Playoff Teams (1962-2001)')
plt.xlabel('Wins')
plt.show()
```

The average playoff team wins about 95 games.  However, the Athletics probably want a greater than average chance to make the playoffs.

If Beane and DePodesta want to put their team in the 70th quantile of playoff teams, they should try to win at least 98 games.  This is of course an imperfect estimate if we were to take into account how well their division rivals do and other factors besides wins, but we can nonetheless establish a concrete benchmark for the team to strive toward - a 98 win regular season.

```python
playoff_teams = teams[teams['Playoffs']==1]
wins = playoff_teams['W'].quantile(.7)
print(wins)
```

## Building a 98-win team

The next piece of analysis for the Athletics management was figuring out how to build a club that could win 98 games.  Winning a game of baseball ultimately boils down to two things: scoring runs and preventing opponents' runs.

These can be quantified by Runs Scored (RS) and Runs Allowed (RA) in the dataset.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='RA', y='RS', hue='Playoffs', data=teams)

plt.title('Runs Scored and Allowed by MLB Teams')
plt.xlabel('Runs Allowed (RA)')
plt.ylabel('Runs Scored (RS)')
plt.show()
```

We can see clearly from the above plot that playoff teams usually score a high amount of runs, and allow a small amount of runs.

RS and RA can be combined into a single statistic known as Run Differential, defined as RS minus RA.  We will add this column to the dataset and attempt to draw similar insights from it.

```python
run_diffs = [row['RS'] - row['RA'] for index, row in teams.iterrows()]
teams['RD'] = run_diffs
```

```python
fig = plt.figure(figsize=(5,7))

sns.stripplot(x='Playoffs', y='RD', hue='Playoffs', data=teams)

plt.title('Run Differential of Playoff and Non-Playoff Teams')
plt.ylabel('Run Differential (RD)')

plt.show()
```

We see again that playoff teams have a higher average run differential, and very few teams make the playoffs with a low or negative run differential.

Beane and DePodesta already have a concrete goal - to win 98 games.  Now the question becomes, how many runs must their team score and allow in order to have a reasonably high probability of winning those 98 games?

For this we will examine the mean team at each number of regular season wins, and compare the runs produced and prevented by each.

```python
group_wins = teams.pivot_table(index='W', values=['RS','RA','RD'], aggfunc=np.mean).astype(int)
group_wins['Wins'] = group_wins.index
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# Axis 1
sns.scatterplot(x='W', y='RD', hue='Playoffs', data=teams, ax=ax1)
ax1.set_title('Run Differential and Games Won')
ax1.set_ylabel('Run Differential (RD)')
ax1.set_xlabel('Wins (W)')
ax1.axvline(98, color='red')
ax1.text(99,-300,'98 win goal', color='red')


# Axis 3
sns.scatterplot(x='Wins', y='RS', data=group_wins, color='green', ax=ax2)
sns.scatterplot(x='Wins', y='RA', data=group_wins, color='purple', ax=ax2)
ax2.set_title('Mean Runs Scored and Allowed by Wins')
ax2.legend(labels=['RS', 'RA'], loc='upper center')
ax2.axvline(98, color='red')

plt.show()
```

We can estimate from this that a team will need to score 771 runs and allow no more than 630 runs in order to have a 98-win season.

In a sense, we don't actually care if a team is better at scoring runs or preventing runs, because the only thing that produces results in a game is run differential.  We see from the above that our team will need to produce a run differential of about 141 runs in order to reach 98 wins and the postseason.

```python
print('RS:', round(teams[teams['W']==98]['RS'].mean(),1))
print('RA:', round(teams[teams['W']==98]['RA'].mean(),1))
print('RD:', round(teams[teams['W']==98]['RD'].mean(),1))
```

## Scoring and Preventing Runs

The next phase of analysis will be to determine what a team must do in order to score 141 runs more than they produce.

Particularly, we are interested in which stat categories correlate most heavily withn run production and prevention.  The Athletics team will then be able to prioritize these statistics more highly than other teams, potentially finding overlooked value-producers in the player market.

The A's did a regression analysis with dozens of stat categories, but for simplicity we will focus on just three major stats - BA, OBP, and SLG.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

Before carrying out a regression analysis, we want to check for linearity among our variables.

```python
stat_colors = {'BA':'royalblue', 'OBP':'goldenrod', 'SLG':'seagreen'}
```

```python
fig = plt.figure(figsize=(9,3))
i=0

for col in ['BA', 'OBP', 'SLG']:
    x = teams.sort_values(col)[[col]].values
    y = teams.sort_values(col)['RS'].values
    
    # generate a best-fit line for each statistic
    poly = PolynomialFeatures(degree=1)
    x_poly = poly.fit_transform(x.reshape(-1,1))
    model = LinearRegression()
    model.fit(x_poly, y)
    y_pred = model.predict(x_poly)
    
    #plot the results
    i+=1
    ax = fig.add_subplot(1,3,i) 
    ax.plot(x, y_pred, c='black')
    ax.scatter(x, y, s=5, c=stat_colors[col])
    
    ax.set_ylim(400,1050)
    ax.set_xlim(0.2,0.5)
    ax.set_title('Runs Scored by ' + col)
    if i==1:
        ax.set_ylabel('Runs Scored')
    ax.set_xlabel(col)

    r2 = r2_score(y, y_pred)
    print(col, r2)
    
plt.show()
```

We see that indeed there seems to be a linear relationship between each of our offensive stat categories and the amount of runs that a team scores.  The individual R-squared values for each category are printed above.

Now that we have satisfied the Linearity condition, we can attempt an OLS regression on the offensive stat categories and Runs Scored.

```python
X = teams[['BA', 'OBP', 'SLG']]
Y = teams['RS']

X = sm.add_constant(X, prepend=False)
RS_model = sm.OLS(Y,X).fit()

print(RS_model.summary())
```

First, we notice that our model has quite a high R-squared value of 0.930, indicating that the model fits the data very well.

Looking at the coefficients of each variable in the table, we see that OBP and SLG are substantially more significant than BA.  This could indicate that we should remove BA from the model altogether.

```python
X = teams[['OBP', 'SLG']]
Y = teams['RS']

X = sm.add_constant(X, prepend=False)
RS_model_2 = sm.OLS(Y,X).fit()

print(RS_model_2.summary())
```

We see that when we remove BA from the model, we lose almost no R-squared value.  This indicates that perhaps our model is better off without BA included, as it does not contribute significantly to the model's fitness.

The final regression we need to run is on the defensive stat categories. For this, we will use OOBP, and OSLG, the opponent's version of the same statistics (ommiting OBA).  We will run this regression with RA (runs allowed) as the Y variable.

Unfortunately, we have much fewer data points for this analysis, because the dataset only tracks OOBP and OSLG from 1999 onward.  However, we will examine the R-squared value with what we do have in order to approximate how strong the model is despite this shortcoming.

```python
X = teams[teams['OOBP'].notnull()][['OOBP', 'OSLG']]
Y = teams[teams['OOBP'].notnull()]['RA']

X = sm.add_constant(X, prepend=False)
RA_model = sm.OLS(Y,X).fit()

print(RA_model.summary())
```

We see that this model still produces a strong R-squared value of 0.907 despite the smaller number of observations.  This regression indicates that OOBP is stronger than OSLG, but that both are statistically significant.

## Achieving a Winning Season

With the above regressions completed, Beane and DePodesta have most of their questions answered.

1. What is our goal for the team?
> To make the playoffs

2. How many wins do we need to make the playoffs?
> 98

3. How many more runs do we need to score than our opponents to get to 98 wins?
> 141

This leaves them with a final question:

4. What OBP and SLG do we need to achieve a run differential of 141?

To answer this question, DePodesta built models to predict the performance of each player on the team.  For our purposes, we will assume that their mostly-intact pitching staff from the previous season would post roughly the same OOBP and OSLG as they had last year.  

The 2001 A's posted a OOBP of 0.308 and a OSLG of 0.380.

Plugging these values into our model, we get an approximate number for runs allowed.

```python
runs_allowed = -837.3779 + 2913.5995*0.308 + 1514.2860*0.380
print('Approximate Runs Allowed:', round(runs_allowed,1))
```

If the A's desire to have a run differential of +141, they would then need to score a number of runs equal to 635+141.

```python
runs_scored = 635 + 141
print('Approximate Runs Scored:', round(runs_scored,1))
```

If we look at the actual outcome of the 2002 season, we find the following results for the A's:
* OBP = 0.339
* SLG = 0.432
* OOBP = 0.315
* OSLG = 0.384

```python
predicted_RS = -804.6271 + 2737.7680*0.339 +  1584.9086*0.432
predicted_RA = -837.3779 + 2913.5995*0.315 + 1514.2860*0.384

print('Predicted Runs Scored:', predicted_RS)
print('Predicted Runs Allowed:', predicted_RA)
```

We can compare this with the actual run figures for the 2002 A's:

* Runs Scored: 800
* Runs Allowed: 654

This is actually a bit better than our desired totals of:

* RS: 771.1
* RA: 630.4

The 2002 A's ended up with 103 wins, placing them 1st in the regular season and making the playoffs.

They would unfortunately lose in the first round to the Minnesota Twins, but the Moneyball approach to baseball analytics soon became the standard across the game.
