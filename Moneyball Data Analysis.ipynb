# Introduction

This analysis uses the 'Moneyball' dataset, which was famously used by Oakland Athletics GM Billy Beane and analyst Paul DePodesta in the early 2000s to revolutionize MLB player evaluation.  In this project, we will attempt to reproduce some of the conclusions by the Athletics club about what drives a team to reach the playoffs.

```python
data = pd.read_csv('baseball.csv')
```

This dataset contains data from 1962-2012. To accurately get a picture of what the 2001-2002 Athletics would have used, we will throw out all the data which is dated past the 2001 season.

```python
teams = data[data['Year']<2002]

teams.head()
```

## Making the Playoffs

Beane and Podesta were interested in dissecting what drives a team to success.  In the MLB, the top teams are placed into a playoff bracket after the conclusion of the regular season, with the winner emerging as the World Series champion.  During this postseason, teams compete in 5-game and 7-game series against one another to determine who advances.  These short series, compared to the 162-game regular season, introduce a high amount of variability and can be quite hard to predict or build around.

With this in mind, Beane and Podesta decided to focus their efforts on building a team which could make the playoffs as their primary goal.  They would then hope for the die to roll in their favor in the postseason games.  

One important piece of this analysis is determining the number of games a team needs to win to give themselves a good chance to make the playoffs.

```python
teams[teams['Playoffs']==1]['W'].mean()
```

We can see that the average playoff team in the dataset won between 95 and 96 games.

To further visualize the variation in playoff team wins, we create the categorical plot below, where the orange dots represent playoff teams and the blue dots represent non-playoff teams.  The vertical black line represents the average number of wins by playoff teams.

```python
sns.catplot(y='Team', x='W', data=teams, jitter=False, hue='Playoffs', height=8)

plt.axvline(teams[teams['Playoffs']==1]['W'].mean(), color='black')
plt.title('Games Won by Playoff and Non-Playoff Teams (1962-2001)')
plt.xlabel('Wins')
plt.show()
```

If Beane and DePodesta can successfully get their team to win 95 games, they can assume that they are about 83% likely to make the playoffs.  So we can use this data to establish a concrete goal for the team that maximizes their chances to secure a playoff berth - a 95 win regular season.

```python
print(round(100*teams[teams['W']>95]['Playoffs'].mean(),1), '%')
```

## Building a 95-win team

The next piece of analysis for the Athletics management was figuring out how to build a club that could win 95 games.  Winning a game of baseball ultimately boils down to two things: scoring runs and preventing opponents' runs.

These can be quantified by Runs Scored (RS) and Runs Allowed (RA) in the dataset.

```python
sns.scatterplot(x='RA', y='RS', hue='Playoffs', data=teams)

plt.title('Runs Scored and Allowed by MLB Teams')
plt.xlabel('Runs Allowed (RA)')
plt.ylabel('Runs Scored (RS)')
plt.show()
```

We can see clearly from the above plot that playoff teams usually score a high amount of runs, and allow a small amount of runs.

RS and RA can be combined into a single statistic known as Run Differential, defined as RS minus RA.  We will add this column to the dataset and attempt to draw similar insights from it.

```python
teams.loc[:,'RD'] = teams['RS'] - teams['RA']
```

```python
sns.stripplot(x='Playoffs', y='RD', hue='Playoffs', data=teams)

plt.title('Run Differential of Playoff and Non-Playoff Teams')
plt.ylabel('Run Differential (RD)')
plt.show()
```

We see again that playoff teams have a higher average run differential, and very few teams make the playoffs with a low or negative run differential.
